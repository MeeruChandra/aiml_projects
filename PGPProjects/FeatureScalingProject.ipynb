## 2. Data cleansing: [15 Marks]

Before any Feautre Engineering, THe data should be split into Train and test so that the test data always represents the real work Production data

```python
#Dropping the Time column as it is only a representation of the time when the observation was recorded

X = df.drop(['Time','Pass/Fail'], axis=1) 
y= df['Pass/Fail']
X.shape
```

```python
from sklearn.model_selection import train_test_split

# Split X and y into training and test set in 80:20 ratio
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=1  , stratify=y, shuffle = True)
# Stratify has been used to ensure equal distribution of churn values in both training and test (we have almost 3 times no of the yes churns
```

### A. Write a for loop which will remove all the features with 20%+ Null values and impute rest with mean of the feature. [5 Marks]

```python
#Check if there are 0 values and replace them with NaN
X_train.eq(0).sum().sort_values( )
#Observation: there are 0 values in many of the colums
```

```python
X_train= X_train.replace(0,np.NaN)
X_test= X_test.replace(0,np.NaN)
```

```python
X_train.eq(0).sum().sort_values( )
#All 0s have been replaced with NaN
```

```python
features_to_drop=[]
for col in X_train.columns:
    # Calculate the percentage of null values for each feature
    null_percentage = (X_train[col].isnull().sum()/X_train.shape[0])*100
    print(f"{col} : {null_percentage}")
    if (null_percentage>20):
        features_to_drop.append(col)
```

```python
len(features_to_drop)
#165 features have 20% or more null values
```

```python
# Drop features with 20% or more null values
X_train.drop(columns=features_to_drop, inplace=True) 
#the same features will be dropped from test set also
X_test.drop(columns=features_to_drop, inplace=True) 
#32 columns were dropped 
```

```python
X_train.shape
#165 features with null values were dropped
```

#### Imputing the rest of the columns to replace nulls with mean value

```python
from sklearn.impute import SimpleImputer

# Impute remaining features with mean
imputer = SimpleImputer(strategy='mean')
X_train_2 = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)
#Trasform the test data will the imputer fitted with the train data set
X_test_2 = pd.DataFrame(imputer.transform(X_test), columns=X_train.columns)
```

### E. Display and explain the classification report in detail. [3 Marks]

```python
from sklearn.metrics import accuracy_score, classification_report ,precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.preprocessing import label_binarize
```

```python
def get_scores(yActual,yPred) :
    # Calculate performance metrics
    accuracy = accuracy_score(yActual, yPred)
    precision = precision_score(yActual, yPred,average='weighted')
    recall = recall_score(yActual, yPred,average='weighted')
    f1 = f1_score(yActual, yPred,average='weighted')
    roc= roc_auc_score(yActual, yPred)
    # Print the performance metrics   
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC AUC Score:{roc:.4f}" )

    print('----------------------------------------------------------')
    print(classification_report(yActual, yPred))
    print('\n')

    cm = metrics.confusion_matrix(yActual, yPred)
    plt.clf()
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Wistia)
    classNames = ['Fail','Pass']
    plt.title('Confusion Matrix - Test Data')
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
    tick_marks = np.arange(len(classNames))
    plt.xticks(tick_marks, classNames, rotation=45)
    plt.yticks(tick_marks, classNames)
    s = [['TN','FP'], ['FN', 'TP']]

    for i in range(2):
        for j in range(2):
            plt.text(j,i, str(s[i][j])+" = "+str(cm[i][j]))
    plt.show()

    
```
