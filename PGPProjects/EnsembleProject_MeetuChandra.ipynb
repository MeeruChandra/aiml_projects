## C. Create a function that will accept a DataFrame as input and return pie-charts for all the appropriate Categorical features. Clearly show percentage distribution in the pie-chart. [4 Marks]

```python
     
#Function to plot Pie-Charts for all categorical variables in the dataframe
def pie_charts_for_CategoricalVar(df,m):
    '''Takes in a dataframe(df_pie) and plots pie charts for all categorical columns.
    m = number of columns required in grid'''
    
    categorical_cols = df.select_dtypes(include=['object']).columns
        
    plt.figure(figsize=(15, 15))
    plt.subplots_adjust(hspace=0.4)
    plt.suptitle("Pie-Charts for Categorical Variables in the dataframe", fontsize=18, y=0.95)
    
    # number of columns, as inputted while calling the function
    ncols = m
    #b=len(categorical_cols)
    # calculate number of rows
    nrows = len(categorical_cols) // ncols + (len(categorical_cols) % ncols > 0)
    
    # loop through the length of 'b' and keep track of index
    for n, i in enumerate(categorical_cols):
        # add a new subplot iteratively using nrows and ncols
        ax = plt.subplot(nrows, ncols, n + 1)

        # filter df and plot 'i' on the new subplot axis
        df.groupby(i).size().plot(kind='pie', autopct='%.2f%%',ax=ax)
        
        ax.set_title(i.upper())
        ax.set_xlabel("")
        ax.set_ylabel("")
    plt.show()
```

### Create functions to get scores, fit and train a model and gets the scores

```python
from sklearn.metrics import accuracy_score, classification_report ,precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn import metrics

def getScores(yActual,yPred, TypeOfData):
    # Calculate performance metrics
    accuracy = accuracy_score(yActual, yPred)
    precision = precision_score(yActual, yPred,average='weighted')
    recall = recall_score(yActual, yPred,average='weighted')
    f1 = f1_score(yActual, yPred,average='weighted')
    roc= roc_auc_score(yActual, yPred)
    # Print the performance metrics   
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC AUC Score:{roc:.4f}" )

    print('----------------------------------------------------------')
    print(classification_report(yActual, yPred))
    print('\n')
    
    cm=metrics.confusion_matrix(yActual, yPred,labels=[0, 1])
    df_cm = pd.DataFrame(cm, index = [i for i in ["No","Yes"]],
                  columns = [i for i in ["No","Yes"]])
    plt.figure(figsize = (5,3))
    sns.heatmap(df_cm, annot=True ,fmt='g')
    plt.title(f"Confusion Matrix for {TypeOfData}")
    plt.ylabel('Prediction')
    plt.xlabel('Actual')    
    plt.show()
    
    return [accuracy, recall, precision, f1,  roc]
    #print('-'*10) 
    
#definiting a function to execute various Models on the data
def FitAndTrainModel(modelType, modelName, modelToFit, XTrain, XTest, yTrain, yTest, resultsArray,saveResults):
    model = modelToFit
    model.fit(XTrain, yTrain)
    ytrain_pred = model.predict(XTrain)
    ytest_pred = model.predict(XTest)
    results=[]
    # Print the performance metrics
    print(f'-------------Perfomance score for {modelName}-------------')
    print(f'-------------Training Dataset-------------')
    train= getScores(yTrain, ytrain_pred, "Training")
    
    print(f'\n-------------Test Dataset-------------') 
    test= getScores(yTest, ytest_pred, "Test")
     
    if saveResults==1:
        resultsArray.append([modelType, modelName,train[0],train[1],train[2],train[3],train[4],test[0],test[1],test[2],test[3],test[4]])
        #resultsArray.append([modelType, modelName,"Train", train[0],train[1],train[2],train[3],train[4]])
        #resultsArray.append([modelType, modelName,"Test", test[0],test[1],test[2],test[3],test[4]])
    return resultsArray
```

## B. Use grid search and improve the performance of the Decision tree model , check the performance of the model on train and test data , provide the differences observed in performance in Q3.a and Q3.b ( 5 marks )

```python
from sklearn.model_selection import GridSearchCV

def performGridSearch(modelToFit,X,y,params, nfolds):    
    #use gridsearch to test all values
    dGridCV = GridSearchCV(modelToFit, params, cv=nfolds)
    #fit model to data
    dGridCV.fit(X, y)
    return dGridCV.best_estimator_
 
```

### As GridSearchCV takes a lot of time and may not work for a large set of parameters, we can use RandomisedSearchCV 

```python
from sklearn.model_selection import RandomizedSearchCV

def performRadomisedSearchCV(modelToFit,X,y,params, nfolds):    
    #use RadomisedSearchCV to test all values    
    dRansomCV = RandomizedSearchCV(estimator = modelToFit, param_distributions = params, n_iter = 100, cv = nfolds,
                           verbose=2, random_state=42, n_jobs = -1)
    
    
    # Fit the random search model
    dRansomCV.fit(X, y)
    return dRansomCV.best_estimator_
```

### Performing Randomised search CV to get the most probable parameters to use

```python
tree_param_grid_random = {
              "max_depth": max_depth,
               "criterion": ["gini", "entropy"],
               'max_leaf_nodes': list(range(2, 100)),
               'min_samples_split': np.arange(3, 100)
               }# Method of selecting samples for training each tree
 
decisionTreeClassifierRandom = DecisionTreeClassifier(random_state=42)
#bestRandomForestClassifier=performGridSearch(randomForestClassifier,X_train, y_train,rf_param_grid, 10)
bestTreeClassifierR = performRadomisedSearchCV(decisionTreeClassifierRandom,X_train, y_train,tree_param_grid_random, 3)
FitAndTrainModel ("DT","DecisionTree with Randomized Search", bestTreeClassifierR,X_train, X_test, y_train, y_test, scoresAll,0 )
#with the best model returned by GridSearch the issue with overfitting on the train data is resolved.
#however the scores are stil not very high
```

```python
def getScoresDataFrame (plotType, results):    
    # Convert results to a DataFrame
    columns = ['ModelType',"Model",'Accuracy_Train', 'Recall_Train',  'Precision_Train', 'F1_Train', 'ROC_Train','Accuracy_Test','Recall_Test', 'Precision_Test',  'F1_Test', 'ROC_Test']
    #columns = ['ModelType',"Model","DataType",'Accuracy', 'Recall',  'Precision', 'F1', 'ROC']
    dfResults =pd.DataFrame(results, columns=columns)
    if plotType != '':
        return dfResults[dfResults['ModelType']==plotType]
    
    return dfResults
    
def PlotResults (nameOfPlot,plotType, results):
    
    # Convert results to a DataFrame
    columns = ['ModelType',"Model",'Accuracy_Train', 'Recall_Train',  'Precision_Train', 'F1_Train', 'ROC_Train','Accuracy_Test','Recall_Test', 'Precision_Test',  'F1_Test', 'ROC_Test']
    #columns = ['ModelType',"Model","DataType",'Accuracy', 'Recall',  'Precision', 'F1', 'ROC']
    dfResults = pd.DataFrame(results, columns=columns)
    if plotType != '':
        results_df = dfResults[dfResults['ModelType']==plotType]
    
    # Plot the results
    plt.figure(figsize=(8, 5))

    metrics = ['Accuracy', 'Recall', 'Precision', 'F1', 'ROC']
    
       
    for metric in metrics:
        plt.plot(results_df['Model'], results_df[f'{metric}_Train'], label=f'Train {metric}')
        plt.plot(results_df['Model'], results_df[f'{metric}_Test'], label=f'Test {metric}', linestyle='dashed')
 
    plt.title('Model Evaluation Metrics')
    plt.xlabel('Model')
    plt.ylabel('Score')
    plt.legend(loc=2, prop={'size': 6})
    plt.xticks(rotation=45, fontsize=8)
    
```

```python
PlotResults ("DecisionTree",'DT', scoresAll)
```
